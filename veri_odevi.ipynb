```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('cereal.csv')
markalar = veriSeti['name']
kalori = veriSeti['calories']

plt.figure(figsize=(20, 7))
plt.bar(markalar, kalori)
plt.xlabel('Markalar')
plt.ylabel('Kalori')
plt.title('Markaların Kalorileri')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('cereal.csv')
markalar = veriSeti['name']
protein = veriSeti['protein']

plt.figure(figsize=(20, 7))
plt.bar(markalar, protein)
plt.xlabel('Markalar')
plt.ylabel('Protein')
plt.title('Markaların Protein Değerleri')
plt.xticks(rotation=90)
plt.show()
```
